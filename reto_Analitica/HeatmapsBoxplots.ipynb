Cargamos el archivo de la base de datos y lo desplegamos a pantalla

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("/content/01.07.2022_da_hospitalizados_covid19 (1).csv")

df1
```

#Gráfica de los datos
Graficamos los datos de pacientes y positivos de la base de datos

```python
fig = plt.figure(figsize=(20,12))

plt.plot(df1['Pacientes'], df1['Positivos'], '*', color = 'red')
plt.xlabel('Pacientes')
plt.ylabel('Positivos')
plt.title('Relación entre cantidad de pacientes y casos positivos de cada institución')
plt.grid(True)
```

#Mapa de calor
Importamos librería **seaborn** y grafícamos el mapa de calor

```python
import seaborn as sns

df1_corr = df1.corr()
sns.heatmap(data=df1_corr, annot=True)
```

#Diagrama de cajas y bigotes
Graficamos diagrama de cajas y bigotes de los pacientes de la base de datos

```python
fig = plt.figure(figsize=(20, 5))

# Gráfico boxplot
sns.boxplot(data=df1, x='Pacientes', color = 'red')

# Ejes y título
plt.title('Histograma de los pacientes por institución')
```

#Gráfico de dispersión

Graficamos los datos de la fecha vs positivos de la base de datos

```python
# Graficaremos la relación entre la fecha y pacientes positivos

# Tamaño de la imagen.
fig = plt.figure(figsize=(20,6))

# Gráfico scatterplot.
sns.scatterplot(data=df1, x ='Fecha', y='Positivos')

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientaciÃƒÂ³n.
plt.title('Relación entre la fecha y pacientes positivos')
plt.xlabel('Fecha')
plt.ylabel('Positivos')
```

#Histograma

```python
fig = plt.figure(figsize=(20,12))

sns.histplot(data=df1, x='Positivos', bins=10, kde=True)
```
